# XGBGridSolver

```python
from IPython.core.display import display
from IPython.core.display import HTML
from src.datamodel.DataModel import Task, Competition
from src.plot import plot_task
from src.solver_multimodel.XGBGridSolver import XGBGridSolver
from src.solver_multimodel.XGBGridSolver import XGBGridSolverDart
from src.solver_multimodel.XGBGridSolver import XGBGridSolverGBtree

%load_ext autoreload
%autoreload 2
```

```python
solver = XGBGridSolver()
```

```python
task   = Task('training/0d3d703e.json')
solver.plot_detects( task )
```

# Tasks

```python
competition = Competition()
competition.map(solver.solve_dataset)
```

```python
print(competition)
```

```python
for name, dataset in competition.items():
    display(HTML("<h2>"+name+"</h2>"))
    for task in dataset:
        if task.score():
            plot_task(task)
```

# Additionally Solved - XGBGridSolverDart

```python
dart_solver = XGBGridSolverDart()
dart_solver.verbose = False
for name, dataset in competition.items():
    display(HTML("<h2>"+name+"</h2>"))
    for task in dataset:
        if not task.score():
            try: dart_solver.solve(task)
            except: pass
            if task.score():
                plot_task(task)
```

# Additionally Solved - XGBGridSolverGBtree

```python
gbtree_solver = XGBGridSolverGBtree()
gbtree_solver.verbose = False
for name, dataset in competition.items():
    display(HTML("<h2>"+name+"</h2>"))
    for task in dataset:
        if not task.score():
            try: gbtree_solver.solve(task)
            except: pass
            if task.score():
                plot_task(task)
```

# Noisy Guesses - XGBGridSolver

```python
for name, dataset in competition.items():
    display(HTML("<h2>"+name+"</h2>"))
    for task in dataset:
        if not task.score():
            solver.solve(task, force=True)
            if not task.score():
                plot_task(task)
```
